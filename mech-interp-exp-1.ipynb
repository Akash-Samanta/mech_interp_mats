```python
import torch
import torch.nn as nn
import numpy as np
from torch.utils.data import DataLoader
import torchvision
import torchvision.transforms as transforms
from sklearn.cluster import KMeans
from scipy.stats import entropy
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.preprocessing import StandardScaler
import seaborn as sns
from scipy.stats import spearmanr
import os
import torchvision.models as models

def create_model():
    model = models.resnet50(pretrained=True)  
    model.eval()  
    return model

def create_data_loaders(batch_size=64):

    normalize = transforms.Normalize(
        mean=[0.485, 0.456, 0.406], 
        std=[0.229, 0.224, 0.225]
    )

    val_transform = transforms.Compose([
        transforms.Resize(256),      
        transforms.CenterCrop(224),  
        transforms.ToTensor(),
        normalize,
    ])

    
    val_dataset = torchvision.datasets.ImageFolder(
        root='/kaggle/input/imagenet1k-val/imagenet-val', 
        transform=val_transform
    )

    val_loader = DataLoader(
        val_dataset, batch_size=batch_size, shuffle=False, 
        num_workers=4, pin_memory=True
    )
    return val_loader



class NeuronImportanceAnalyzer:
    def __init__(self, model, device, trainloader, testloader):
        self.device = device
        self.trainloader = trainloader
        self.testloader = testloader
        self.activation_maps = {}
        self.gradient_maps = {}
        
        # use base model, not DataParallel
        self.model = model.module if isinstance(model, nn.DataParallel) else model
        self.model.eval()
        self.register_hooks()

    def register_hooks(self):
        def get_activation(name):
            def hook(module, input, output):
                self.activation_maps[name] = output.detach()
            return hook

        def get_gradient(name):
            def hook(module, grad_input, grad_output):
                self.gradient_maps[name] = grad_output[0].detach()
            return hook


        for name, module in self.model.named_modules():
            if isinstance(module, nn.Conv2d):
                module.register_forward_hook(get_activation(name))
                module.register_full_backward_hook(get_gradient(name))

    def compute_gradient_importance(self, num_batches=None):
        gradient_scores = {}
        
        for batch_idx, (inputs, targets) in enumerate(tqdm(self.trainloader, desc="Computing gradient importance")):
            if num_batches and batch_idx >= num_batches:
                break
                
            inputs, targets = inputs.to(self.device), targets.to(self.device)
            
            self.model.zero_grad()
            outputs = self.model(inputs)
            loss = nn.CrossEntropyLoss()(outputs, targets)
            loss.backward()
            
            for name, gradients in self.gradient_maps.items():
                if name not in gradient_scores:
                    gradient_scores[name] = []
                # Keep computation on GPU
                grad_norm = torch.norm(gradients, p=2, dim=(2,3)).mean(0)
                gradient_scores[name].append(grad_norm)
        
    
        return {name: torch.stack(scores).mean(0).cpu().numpy() 
                for name, scores in gradient_scores.items()}

    def compute_variance_importance(self, num_batches=None):
        variance_scores = {}
        
        with torch.no_grad():
            for batch_idx, (inputs, _) in enumerate(tqdm(self.trainloader, desc="Computing variance importance")):
                if num_batches and batch_idx >= num_batches:
                    break
                    
                inputs = inputs.to(self.device)
                _ = self.model(inputs)
                
                for name, activations in self.activation_maps.items():
                    if name not in variance_scores:
                        variance_scores[name] = []
                    
                    # Keep all computations on GPU
                    mean = activations.mean(dim=1, keepdim=True)
                    variance = (activations - mean).pow(2)
                    channel_variance = variance.mean(dim=(0,2,3))
                    variance_scores[name].append(channel_variance)
        
        
        return {name: torch.stack(vars).mean(0).cpu().numpy() 
                for name, vars in variance_scores.items()}

    def compute_tei_importance(self, num_batches=None):
        tei_scores = {}
        
        for batch_idx, (inputs, targets) in enumerate(tqdm(self.trainloader, desc="Computing TEI")):
            if num_batches and batch_idx >= num_batches:
                break
                
            inputs, targets = inputs.to(self.device), targets.to(self.device)
            
            self.model.zero_grad()
            outputs = self.model(inputs)
            loss = nn.CrossEntropyLoss()(outputs, targets)
            loss.backward()
            
            for name in self.activation_maps:
                if name not in tei_scores:
                    tei_scores[name] = []
                
                # Compute TEI on GPU: |a_j * L_j|
                activations = self.activation_maps[name]
                gradients = self.gradient_maps[name]
                
            
                tei = torch.abs(activations * gradients)
            
                tei_score = tei.mean(dim=(0,2,3))
                tei_scores[name].append(tei_score)
       
        return {name: torch.stack(scores).mean(0).cpu().numpy() 
                for name, scores in tei_scores.items()}

    def compute_all_importance_metrics(self, num_batches=None):
        print("Computing gradient and variance metrics...")
        
        gradient_scores = {}
        variance_scores = {}
        
        for batch_idx, (inputs, targets) in enumerate(tqdm(self.trainloader, desc="Computing metrics")):
            if num_batches and batch_idx >= num_batches:
                break
                
            inputs, targets = inputs.to(self.device), targets.to(self.device)
            
            # Forward pass
            self.model.zero_grad()
            outputs = self.model(inputs)
            loss = nn.CrossEntropyLoss()(outputs, targets)
            loss.backward()
            

            for name in self.activation_maps:
                activations = self.activation_maps[name]
                gradients = self.gradient_maps[name]
                
                if name not in gradient_scores:
                    gradient_scores[name] = []
                    variance_scores[name] = []
                
                # 1. Gradient importance
                grad_norm = torch.norm(gradients, p=2, dim=(2,3)).mean(0)
                gradient_scores[name].append(grad_norm)
                
                # 2. Variance importance
                mean = activations.mean(dim=1, keepdim=True)
                variance = (activations - mean).pow(2)
                channel_variance = variance.mean(dim=(0,2,3))
                variance_scores[name].append(channel_variance)
        
        # Convert to numpy arrays 
        results = {
            'gradient': {name: torch.stack(scores).mean(0).cpu().numpy() 
                        for name, scores in gradient_scores.items()},
            'variance': {name: torch.stack(scores).mean(0).cpu().numpy() 
                        for name, scores in variance_scores.items()}
        }
        
        return results
    
    def compute_unified_importance_score(self, importance_metrics=None):
        """Compute unified score using gradient and variance metrics"""
        if importance_metrics is None:
            importance_metrics = self.compute_all_importance_metrics()
        
        unified_scores = {}
        for name in importance_metrics['gradient']:

            grad_tensor = torch.tensor(importance_metrics['gradient'][name], device=self.device)
            var_tensor = torch.tensor(importance_metrics['variance'][name], device=self.device)
            
            # Normalize 
            def normalize_tensor(x):
                mean = x.mean()
                std = x.std()
                return (x - mean) / (std + 1e-8)
            
            grad_norm = normalize_tensor(grad_tensor)
            var_norm = normalize_tensor(var_tensor)
            
            # Combine scores on GPU (average of gradient and variance)
            unified = (grad_norm + var_norm) / 2
            unified_scores[name] = unified.cpu().numpy()
        
        return unified_scores
    
    def perform_ablation_test(self, layer_name, original_acc, metrics=None, percent=10):
        """Test accuracy when ablating top k% neurons using unified importance scores"""
        
        print(f"Original accuracy: {original_acc:.2f}%")
        
        # Get unified importance scores
        unified_scores = self.compute_unified_importance_score(metrics)[layer_name]
        
        # Calculate number of neurons to ablate
        num_neurons = len(unified_scores)
        k = max(1, int(num_neurons * (percent/100)))
        
        # Get top k% neurons based on unified scores
        top_neurons = np.argsort(unified_scores)[-k:]
        print(f"Ablating {k} neurons ({percent}% of {num_neurons} total neurons)")
        
        def ablation_hook(module, input, output):
            mask = torch.ones_like(output, device=self.device)
            mask[:, top_neurons] = 0
            return output * mask
        
        # Register temporary hook
        handle = None
        for name, module in self.model.named_modules():
            if name == layer_name:
                handle = module.register_forward_hook(ablation_hook)
                break
        
        if handle is None:
            raise ValueError(f"Layer {layer_name} not found in model")
        
        ablated_acc = self.evaluate()
        print(f"Accuracy after ablating top {percent}% neurons: {ablated_acc:.2f}%")
        print(f"Accuracy drop: {original_acc - ablated_acc:.2f}%")
        
        handle.remove()
        return original_acc - ablated_acc



    
    def evaluate(self):
        correct = 0
        total = 0
        
        self.model.eval()
        with torch.no_grad():
            for inputs, targets in tqdm(self.testloader, desc="Evaluating"):
                inputs, targets = inputs.to(self.device), targets.to(self.device)
                outputs = self.model(inputs)
                _, predicted = outputs.max(1)
                total += targets.size(0)
                correct += predicted.eq(targets).sum().item()
        
        return 100. * correct / total

    def visualize_importance_distributions(self, layer_name, metrics=None, save_path=None):
        if metrics is None:
            metrics = self.compute_all_importance_metrics()
            
        gradient_scores = metrics['gradient'][layer_name]
        variance_scores = metrics['variance'][layer_name]
        unified_scores = self.compute_unified_importance_score(metrics)[layer_name]
        
        plt.figure(figsize=(12, 4))
        
        plt.subplot(1,3,1)
        sns.histplot(gradient_scores, kde=True)
        plt.title('Gradient-based Importance')
        
        plt.subplot(1,3,2)
        sns.histplot(variance_scores, kde=True)
        plt.title('Variance-based Importance')
        
        plt.subplot(1,3,3)
        sns.histplot(unified_scores, kde=True)
        plt.title('Unified Importance Score\n(Gradient + Variance)')
        
        plt.tight_layout()
        
        if save_path:
            plt.savefig(save_path)
        plt.show()

def main():
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")
    
    model = create_model()
    model = model.to(device)
    
    val_loader = create_data_loaders()
    analyzer = NeuronImportanceAnalyzer(model, device, val_loader, val_loader)

    original_accuracy = analyzer.evaluate()
    print(f"Original accuracy: {original_accuracy:.2f}%")

    os.makedirs("analysis_results", exist_ok=True)

    metrics = analyzer.compute_all_importance_metrics()
    
    # Analysis for each conv layer
    for name, module in model.named_modules():
        if isinstance(module, nn.Conv2d):
            print(f"\nAnalyzing layer: {name}")
            
            # Create output directory for this layer
            os.makedirs(f"analysis_results/{name}", exist_ok=True)
            
            np.save(f"analysis_results/{name}/gradient_scores.npy", 
                    metrics['gradient'][name])
            np.save(f"analysis_results/{name}/variance_scores.npy", 
                    metrics['variance'][name])
            np.save(f"analysis_results/{name}/unified_scores.npy", 
                    analyzer.compute_unified_importance_score(metrics)[name])
            
            # Visualize distributions 
            analyzer.visualize_importance_distributions(
                name,
                metrics=metrics,
                save_path=f"analysis_results/{name}/importance_distributions.png"
            )
            
            # Perform ablation test 
            ablation_impact = analyzer.perform_ablation_test(
                name, 
                original_accuracy,
                metrics=metrics 
            )
            
            with open(f"analysis_results/{name}/summary.txt", "w") as f:
                f.write(f"Ablation impact: {ablation_impact:.2f}%\n")

if __name__ == "__main__":
    main()
```
